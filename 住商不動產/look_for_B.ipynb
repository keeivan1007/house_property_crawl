```python
from hourse_house import *
from selenium import webdriver
from bs4 import BeautifulSoup
import time


chrome_path = "C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe" #chromedriver.exe執行檔所存在的路徑
web = webdriver.Chrome(chrome_path)

#控制板(區)
file_xlsx =  '住商不動產-基隆(1460).xlsx' #輸入要爬取的檔案路近
target_page = 74  #目標對象總頁數
url_1_page = """http://www.hbhousing.com.tw/buyHouse/?q=2^1^2^002^^P0_0^^^0_0^0_0^0^0_0^0_0^0^0^0^0^^6^1^1^1"""
#目標網址 從第一頁開始 ↑

total_page = 0  #目前在爬的頁數進度
crawal_page = target_page - total_page #要爬的多少頁
crawal_count = 0 #爬取的多少頁
error = 0

web.get(url_1_page)
soup = BeautifulSoup(web.page_source, 'xml')
total_list = crawal_soup(soup)
save_excel_House(total_list,file_xlsx)
crawal_count = crawal_count + 1
total_page = total_page + 1

web.find_element_by_xpath('//*[@id="pagewrap"]/span[9]').click() #第一頁特別處理 因為按鈕關係
print('{} page is finish'.format(total_page))

while 1 <= crawal_page:
    
    try:
        soup = BeautifulSoup(web.page_source, 'xml')
        total_list = crawal_soup(soup)
        save_excel_House(total_list,file_xlsx)
        web.find_element_by_xpath('//*[@id="pagewrap"]/span[11]').click()
        crawal_count = crawal_count+1
        total_page = total_page + 1
        print('{} page is finish'.format(total_page))
        
    except Exception as ex:
        print(ex)
        print('We have problem in {} page',format(total_page))
        web.find_element_by_xpath('//*[@id="pagewrap"]/span[11]').click()
        crawal_count = crawal_count+1
        total_page = total_page + 1

    
    time.sleep(3)


print('We have {0} finish,{1} errors'.format(crawal_count-1,error))
#web.close()
web.quit() #關閉整個瀏覽器
```
